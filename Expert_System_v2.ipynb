## XTB API

Conexión con la API del broker XTB que permite la ejecutar ordenes de compra/venta, revisar estado de cuenta y consultar información del mercado por medio del parametro command.

```python
import json
import websocket


from websocket import create_connection
from websocket._exceptions import WebSocketConnectionClosedException
from typing import Any, Dict, Callable


class Client:
	__user_id: str = None
	__password: str = None
	mode: str = None
	__stream_session_id: str = None
	__ws: websocket = None

	def __get_data(self, command: str, **parameters) -> dict:
		data: Dict[str, str] = {
			"command": command,
		}
		if parameters:
			data['arguments'] = {}
			for (key, value) in parameters.items():
				data['arguments'][key] = value
		return data

	def __send(self, data: dict) -> dict:
		self.__ws.send(json.dumps(data))

		response: Any = self.__ws.recv()
		response: Any = json.loads(response)

		if response['status'] == True:
			if 'returnData' in response.keys():
				return response['returnData']
			else:
				return response
		else:
			raise Exception(response)

	def __connect(self) -> None:
		try:
			self.__ws = create_connection(f'wss://ws.xtb.com/{self.mode}')

			data: dict = self.__get_data(
				command='login', userId=self.__user_id, password=self.__password)

			response: dict = self.__send(data)

			self.__stream_session_id = response['streamSessionId']
		except Exception as e:
			raise e

	def __verify__connection(function: Callable) -> Any:
		def wrapper(self, *args, **kwargs):
			if self.__ws == None:
				raise Exception('Please login first with the login method.')
			else:
				try:
					result: Any = function(self, *args, **kwargs)
					return result
				except WebSocketConnectionClosedException:
					self.__connect()
					result: Any = function(self, *args, **kwargs)
					return result
				except BrokenPipeError:
					self.__connect()
					result: Any = function(self, *args, **kwargs)
					return result
				except Exception as e:
					raise e
		return wrapper

	def login(self, user_id: str, password: str, mode: str = 'demo') -> None:
		self.__user_id = user_id
		self.__password = password
		self.mode = mode

		self.__connect()

	def logout(self) -> None:
		return self.exec_command(command='logout')

	@property
	def stream_session_id(self):
		self.exec_command(command='ping')
		return self.__stream_session_id

	@__verify__connection
	def exec_command(self, command: str, **kwargs) -> dict:
		data: dict = self.__get_data(command=command, **kwargs)

		return self.__send(data)
```

## Data Processing

```python
x_columns: List[str] = [
    'open',
    'close',
    'high',
    'low',
    'volume',
    'year',
    'month',
    'week',
    'weekday',
    'day',
    'hour',
    'minute',
    # 'SMA',
    # 'WMA',
    # 'EMA',
    # 'STOCH'
]

y_columns: List[str] = ['y_high', 'y_low']

def data_preprocessing(candles: List[dict]) -> pd.DataFrame:
    candles = pd.DataFrame(candles)
    
    candles[['open', 'close', 'high', 'low']] = candles[['open', 'close', 'high', 'low']] / 100000

    candles['close'] += candles['open']
    candles['high'] += candles['open']
    candles['low'] += candles['open']

    candles = candles.rename(
        columns={
            'ctm': 'c_timestamp',
            'ctmString': 'timestamp',
            'vol': 'volume'
        }
    )

    candles['c_timestamp'] = candles['c_timestamp'] / 1000
    
    # Convert broker timezone to Colombian timezone
    candles['timestamp'] = candles['c_timestamp'].apply(
        lambda t: datetime.fromtimestamp(t, timezone('America/Bogota'))
    )

    # Delete duplicated values
    candles = candles[~candles.duplicated()]

    # Datetime disintegration
    candles[[
        'year',
        'month',
        'week',
        'weekday',
        'day',
        'hour',
        'minute'
    ]] = candles['timestamp'].apply(
        lambda time: pd.Series([
            time.year,
            time.month,
            time.isocalendar()[1],
            time.weekday(), time.day,
            time.hour,
            time.minute
        ])
    )

    # Simple Moving Average Indicator
    candles['SMA'] = TA.SMA(candles)

    # Weighted Moving Average Indicator
    candles['WMA'] = TA.WMA(candles)

    # Exponential Moving Average Indicator
    candles['EMA'] = TA.EMA(candles)

    # Stochastic Indicator
    candles['STOCH'] = TA.STOCH(candles)

    # Delete null values
    candles = candles.dropna()

    return candles

def data_multiplier(candles: pd.DataFrame, multiplier: int = 1) -> pd.DataFrame:
    candles_multiplier: pd.DataFrame = pd.DataFrame()

    for i in range(candles.shape[0]):
        end_datetime: Timestamp = candles.iloc[i]['timestamp'] - timedelta(minutes=1)
        start_datetime: Timestamp = end_datetime - timedelta(minutes=multiplier) + timedelta(minutes=1)
        
        last_hour_candles = candles[candles['timestamp'].between(start_datetime, end_datetime)].reset_index(drop=True)

        v_control: int = last_hour_candles['c_timestamp'].count() == multiplier

        if v_control:
            first_row: pd.Series = last_hour_candles.iloc[0]
            last_row: pd.Series = last_hour_candles.iloc[-1]

            x: pd.DataFrame = pd.DataFrame([first_row])

            x['close'] = last_row['close']
            x['high'] = last_hour_candles['high'].max()
            x['low'] = last_hour_candles['low'].min()
            x['volume'] = last_hour_candles['volume'].sum()
            # x['SMA'] = last_row['SMA']
            # x['WMA'] = last_row['WMA']
            # x['EMA'] = last_row['EMA']
            # x['STOCH'] = last_row['STOCH']

            candles_multiplier = pd.concat([candles_multiplier, x])

    candles_multiplier = candles_multiplier.reset_index(drop=True)

    candles_multiplier['y_timestamp'] = candles_multiplier['timestamp'] + timedelta(minutes=multiplier)

    return candles_multiplier.reset_index(drop=True)

def data_outputs(candles: pd.DataFrame, candles_multiplier: pd.DataFrame, future: int = 0, train: bool=True) -> pd.DataFrame:   
    candles_output: pd.DataFrame = pd.DataFrame()

    if train:
        for i in range(candles_multiplier.shape[0]):
            start_datetime: Timestamp = candles_multiplier.iloc[i]['y_timestamp']
            end_datetime: Timestamp = start_datetime + timedelta(minutes=future) - timedelta(minutes=1)
            
            future_candles = candles[candles['timestamp'].between(start_datetime, end_datetime)].reset_index(drop=True)
            
            v_control: int = future_candles['c_timestamp'].count() == future
            
            if v_control:
                row: pd.Series = candles_multiplier.iloc[i]

                df: pd.DataFrame = pd.DataFrame([row])

                df['y_high'] = (future_candles['high'].max() - df['close']) / BPS
                df['y_low'] = (df['close'] - future_candles['low'].min()) / BPS

                candles_output = pd.concat([candles_output, df])

            candles_output = candles_output.reset_index(drop=True)

        return candles_output
    else:
        candles_multiplier['y_high'] = pd.NaT
        candles_multiplier['y_low'] = pd.NaT

        return candles_multiplier

def _data_outputs(candles: pd.DataFrame, train: bool=True) -> pd.DataFrame:
    
    if train:
        tmp = pd.DataFrame()

        for i in candles.index:
            tmp_datetime = candles.loc[i]['timestamp']
            tmp_candle = candles[candles['timestamp'] == tmp_datetime + timedelta(minutes=1)][['timestamp', 'high', 'low']]
            
            if tmp_candle.size != 0:
                tmp_candle = tmp_candle.set_index(pd.Index([i]))
                tmp_candle = tmp_candle.rename(
                    columns={
                        'timestamp': 'y_timestamp',
                        'high': 'y_high',
                        'low': 'y_low'
                    }
                )
                tmp = pd.concat([tmp, tmp_candle])

        candles = pd.concat([candles,tmp], axis=1)

        candles = candles.dropna()
    else:
        candles['y_high'] = pd.NaT
        candles['y_low'] = pd.NaT
    
    return candles

def data_normalize(candles_x: pd.DataFrame, candles_y: pd.DataFrame, train: bool=True, sc_x:MinMaxScaler = None, sc_y:MinMaxScaler = None) -> tuple:
    if train:
        sc_x = MinMaxScaler(feature_range=(0,1))
        sc_y = MinMaxScaler(feature_range=(0,1))

        candles_n_x = sc_x.fit_transform(candles_x)
        candles_n_y = sc_y.fit_transform(candles_y)
    else:
        if sc_x == None or sc_y == None:
            raise Exception('Please, add sc_x and sc_y arguments.')
        else:
            candles_n_x = sc_x.transform(candles_x)
            candles_n_y = sc_y.transform(candles_y)
    
    return candles_n_x, candles_n_y, sc_x, sc_y

def data_history(candles_x: np.array, candles_y: np.array, history: int = 0):
        candles_h_x = []
        candles_h_y = []
        m = len(candles_x)

        candles_m_x: list = []
        
        for i in range(history,m):
            candles_h_x.append(candles_x[i-history:i])
            candles_m_x.append(candles_x[i])

            candles_h_y.append(candles_y[i])

        candles_h_x, candles_h_y = np.array(candles_h_x), np.array(candles_h_y)

        return candles_h_x, candles_h_y, candles_m_x

def data_processing_mlp(candles: List[dict], multiplier: int = 1, future: int = 10, train: bool = True, sc_x:MinMaxScaler = None, sc_y:MinMaxScaler = None) -> tuple:
    candles: pd.DataFrame = data_preprocessing(candles)
    candles_multiplier: pd.DataFrame = data_multiplier(candles, multiplier)
    candles_outputs = data_outputs(candles, candles_multiplier, future, train)
    candles_normalize_x, candles_normalize_y, sc_x, sc_y = data_normalize(candles_outputs[x_columns], candles_outputs[y_columns], train, sc_x, sc_y)

    return candles, candles_normalize_x, candles_normalize_y, sc_x, sc_y

def data_processing_lstm(candles: List[dict], multiplier: int = 1, future: int = 10, history: int = 60, train: bool = True, sc_x:MinMaxScaler = None, sc_y:MinMaxScaler = None):
    candles: pd.DataFrame = data_preprocessing(candles)
    candles_multiplier: pd.DataFrame = data_multiplier(candles, multiplier)
    candles_outputs = data_outputs(candles, candles_multiplier, future, train)
    candles_normalize_x, candles_normalize_y, sc_x, sc_y = data_normalize(candles_outputs[x_columns], candles_outputs[y_columns], train, sc_x, sc_y)
    candles_history_x, candles_history_y, candles_m_x = data_history(candles_normalize_x, candles_normalize_y, history)

    return candles, candles_m_x, candles_history_x, candles_history_y, sc_x, sc_y
```

## Decision Maker

```python
def decision_maker(model: Sequential, model_type: str, sc_x: MinMaxScaler, sc_y: MinMaxScaler, params: dict, period: int, symbol: str, bps: float, x_columns: list, y_columns: list, multiplier: int = 1, future: int = 10, history: int = 60) -> pd.DataFrame:

    # Data Download
    end: datetime = datetime.now(timezone("America/Bogota"))

    start: datetime = end - timedelta(**{'hours': params['simulation_hours']})
    start: float = start.timestamp() * 1000

    info: dict = {
        "period": period,
        "start": start,
        "symbol": symbol
    }

    last_candles = client.exec_command(command='getChartLastRequest', info=info)

    last_candles = {'rateInfos': pd.read_csv('/content/drive/MyDrive/tgmioe/Archivos descargados/candles.csv')}

    tmp = pd.DataFrame(last_candles['rateInfos'])
    tmp
    tmp[tmp['ctmString']=='Nov 17, 2022, 12:00:00 AM']
    #tmp = tmp[tmp['ctm']>=1668639600000]
    tmp = tmp[tmp['ctm']>=1670799600000]
    tmp[tmp['ctmString']=='Dec 20, 2022, 7:51:00 PM']
    last_candles = {'rateInfos': tmp[tmp['ctm']<=1671562260000]}

    # Data Processing
    candles: pd.DataFrame = None
    x_candles: np.array = None
    y_candles: np.array = None

    if model_type == 'MLP':
        candles, x_candles, y_candles, sc_x, sc_y = data_processing_mlp(
            candles=last_candles['rateInfos'],
            multiplier=multiplier,
            future=future,
            train=False,
            sc_x=sc_x,
            sc_y=sc_y
        )
        # Prediction
        y_predict: np.ndarray = model.predict(x_candles)

        y_predict: pd.DataFrame = pd.DataFrame(sc_y.inverse_transform(y_predict), columns=y_columns)

        # Variables Calculate
        predict: pd.DataFrame = pd.DataFrame(sc_x.inverse_transform(x_candles), columns=x_columns)
    elif model_type == 'LSTM':
        candles, x_candles_c, x_candles, y_candles, sc_x, sc_y = data_processing_lstm(
            candles=last_candles['rateInfos'],
            multiplier=multiplier,
            history=history,
            future=future,
            train=False,
            sc_x=sc_x,
            sc_y=sc_y
        )
        # Prediction
        y_predict: np.ndarray = model.predict(x_candles)

        y_predict: pd.DataFrame = pd.DataFrame(sc_y.inverse_transform(y_predict), columns=y_columns)

        # Variables Calculate
        predict: pd.DataFrame = pd.DataFrame(sc_x.inverse_transform(x_candles_c), columns=x_columns)
    
    # Timestamp Calculate
    predict['timestamp'] = predict.apply(
        lambda row: Timestamp(
            year=round(row.year),
            month=round(row.month),
            day=round(row.day),
            hour=round(row.hour),
            minute=round(row.minute),
            tz=timezone('America/Bogota')
        ), axis=1
    )

    predict['y_timestamp'] = predict['timestamp'] + timedelta(minutes=multiplier)

    bid: float = predict['close']
    spred: float = 0.0001
    ask: float = bid + spred

    predict['bid'] = bid
    predict['diff_h_v'] = y_predict['y_high']
    predict['diff_l_v'] = y_predict['y_low']
    predict['y_high_p'] = predict['bid'] + predict['diff_h_v'] * bps
    predict['y_low_p'] = predict['bid'] - predict['diff_l_v'] * bps
    predict['y_c_high_p'] = np.where(predict['diff_h_v'] < params['loss_min'] + spred/bps, predict['bid'] + ((params['loss_min'] + spred/bps) * bps), predict['y_high_p'])
    predict['y_c_low_p'] = np.where(predict['diff_l_v'] < params['loss_min'] + spred/bps, predict['bid'] - (params['loss_min'] * bps), predict['y_low_p'])
    predict['diff_h'] = ( predict['y_c_high_p'] - predict['bid'] ) / bps
    predict['diff_l'] = ( predict['bid'] - predict['y_c_low_p'] ) / bps
    predict['op'] = np.where(predict['diff_h'] > predict['diff_l'], 'c', 'v')
    predict['price'] = np.where(predict['op'] == 'c', ask, bid)
    predict['take_profit'] = np.where(predict['op'] == 'c', predict['y_c_high_p'] - params['profit_interval']*bps, predict['y_c_low_p'] + params['profit_interval']*bps)
    predict['stop_loss'] = np.where(predict['op'] == 'c', predict['y_c_low_p'] - params['loss_interval']*bps, predict['y_c_high_p'] + params['loss_interval']*bps)
    predict['profit'] = np.where(predict['op'] == 'c', predict['take_profit'] - predict['price'], predict['price'] - predict['take_profit']) / bps
    predict['loss'] = np.where(predict['op'] == 'c', predict['stop_loss'] - predict['price'], predict['price'] - predict['stop_loss']) / bps
    predict['risk'] = predict['profit'] / (-predict['loss'])

    # Decision
    predict['v1'] = predict['profit'] > params['profit_min']
    predict['v2'] = predict['risk'] > params['risk']
    predict['v3'] = True
    predict['v4'] = predict.apply(lambda row: row['y_timestamp'].hour in params['trade_hours'], axis=1)
    predict['v5'] =  np.where(predict['diff_h_v'] > 0, True, False)
    predict['v6'] =  np.where(predict['diff_l_v'] > 0, True, False)
    predict['vf'] = predict['v1'] & predict['v2'] & predict['v3'] & predict['v4'] & predict['v5'] & predict['v6']

    return candles, predict
```

## Backtesting

Simulación compuesta de dos clases:


*   Order: Abstrae el comportamiento de una orden de compra/venta en el mercado de divisas.
*   Account: Abstrae el comportamiento de una cuenta con un Broker para operar en el mercado de divisas.



```python
class Order:
    # Constant
    __spreed: int = 0.0001

    # Init
    __id: int = None
    __open_date: Timestamp = None
    __symbol: str = None
    __cmd: int = None
    __volume: float = None
    __open_price: float = None
    __sl: float = None
    __tp: float = None
    __type_cmd: int = None
    __leverage: float = None
    __batch_size: float = None
    __margin: float = None

    # Finally
    __close_date: Timestamp = None
    __close_price: float = None

    # Cost
    __rollover: float = 0.0
    __commission: float = 0.0
    __swap: float = 0.0

    # Profit
    __profit: float = 0.0

    # State
    __state: str = 'pending'

    # Current Candles
    __current_candle: Dict[str, float] = {
        'timestamp': None,
        'open': None,
        'high': None,
        'low': None,
        'close': None
    }

    @property
    def spread(self):
        return self.__spread

    @property
    def id(self):
        return self.__id

    @property
    def open_date(self):
        return self.__open_date

    @property
    def symbol(self):
        return self.__symbol
        
    @property
    def cmd(self):
        return self.__cmd
        
    @property
    def volume(self):
        return self.__volume

    @property
    def open_price(self):
        return self.__open_price

    @property
    def sl(self):
        return self.__sl

    @property
    def tp(self):
        return self.__tp

    @property
    def type_cmd(self):
        return self.__type_cmd

    @property
    def leverage(self):
        return self.__leverage
        
    @property
    def batch_size(self):
        return self.__batch_size
        
    @property
    def margin(self):
        return self.__margin

    @property
    def close_date(self):
        return self.__close_date

    @property
    def close_price(self):
        return self.__close_price
        
    @property
    def rollover(self):
        return self.__rollover
        
    @property
    def commission(self):
        return self.__commission
        
    @property
    def swap(self):
        return self.__swap
    
    @property
    def profit(self):
        return self.__profit
        
    @property
    def state(self):
        return self.__state
        
    @property
    def current_candle(self):
        return self.__current_candle

    @current_candle.setter
    def current_candle(self, current_candle: Dict[str, float]):
        self.__current_candle = current_candle

    def __init__(self, open_date: Timestamp, symbol: str, cmd: int, volume: float, sl: float, tp: float, type_cmd: int = 0, leverage: float = 200.0, batch_size: int = 100000.0):
        self.__id = open_date.strftime('%Y%m%d%H%M%S%f')
        self.__open_date = open_date
        self.__symbol = symbol
        self.__cmd = cmd
        self.__volume = volume
        self.__sl = sl
        self.__tp = tp
        self.__type_cmd = type_cmd
        self.__leverage = leverage
        self.__batch_size = batch_size

    def update_state(self):
        if self.__state == 'pending':
            # Change state
            if self.__current_candle['timestamp'] >= self.__open_date:
                self.__state = 'open'
                self.__open_price = self.__current_candle['open'] + self.__spreed if self.__cmd == 0 else self.__current_candle['open']
                self.__margin = self.__batch_size * self.__volume * self.__open_price / self.__leverage
                self.__profit = (self.__current_candle['open'] - self.__open_price if self.__cmd == 0 else self.__open_price - self.__current_candle['open']) * self.__batch_size * self.__volume
        if self.__state == 'open':
            if self.__cmd == 0:
                if self.__sl >= self.__current_candle['low']:
                    self.__state = 'close'
                    self.__close_date = self.__current_candle['timestamp']
                    self.__close_price = self.__sl
                    self.__profit = (self.__close_price - self.__open_price) * self.__batch_size * self.__volume
                elif self.__tp <= self.__current_candle['high']:
                    self.__state = 'close'
                    self.__close_date = self.__current_candle['timestamp']
                    self.__close_price = self.__tp
                    self.__profit = (self.__close_price - self.__open_price) * self.__batch_size * self.__volume
                else:
                    self.__profit = (self.__current_candle['open'] - self.__open_price) * self.__batch_size * self.__volume
            else:
                if self.__sl <= self.__current_candle['high']:
                    self.__state = 'close'
                    self.__close_date = self.__current_candle['timestamp']
                    self.__close_price = self.__sl
                    self.__profit = (self.__open_price - self.__close_price) * self.__batch_size * self.__volume
                elif self.__tp >= self.__current_candle['low']:
                    self.__state = 'close'
                    self.__close_date = self.__current_candle['timestamp']
                    self.__close_price = self.__tp
                    self.__profit = (self.__open_price - self.__close_price) * self.__batch_size * self.__volume
                else:
                    self.__profit = (self.__open_price - self.__current_candle['open']) * self.__batch_size * self.__volume
        if self.__state == 'close':
            pass

    def get_state(self) -> Dict[str, Any]:
        return {
            # Init
            'id': self.__id,
            'open_date': self.__open_date,
            'symbol': self.__symbol,
            'cmd': self.__cmd,
            'volume': self.__volume,
            'open_price': self.__open_price,
            'sl': self.__sl,
            'tp': self.__tp,
            'type_cmd': self.__type_cmd,
            'leverage': self.__leverage,
            'batch_size': self.__batch_size,
            'margin': self.__margin,

            # Finally
            'close_date': self.__close_date,
            'close_price': self.__close_price,

            # Cost
            'rollover': self.__rollover,
            'commission': self.__commission,
            'swap' : self.__swap,

            # Profit
            'profit': self.__profit,

            # State
            'state': self.__state,
            'current_candle': self.__current_candle
        }


class Account:
    # Init
    __id: int = None
    __deposits: list = [].copy()
    __open_date: Timestamp = None

    # Summary
    __heritage: float = 0
    __balance: float = None
    __free_margin: float = 0
    __margin: float = 0
    __profit: float = 0
    __margin_level: float = 0

    # Orders
    __orders: Dict[str, Dict[str, Order]] = {
        'pending': {},
        'open': {},
        'close': {}
    }

    # Current Candles
    __current_candle: Dict[str, float] = {
        'timestamp': None,
        'open': None,
        'high': None,
        'low': None,
        'close': None
    }

    @property
    def deposits(self):
        return self.__deposits

    @property
    def id(self):
        return self.__id
    
    @property
    def balance(self):
        return self.__balance

    @property
    def open_date(self):
        return self.__open_date
    
    @property
    def heritage(self):
        return self.__heritage
    
    @property
    def free_margin(self):
        return self.__free_margin

    @property
    def margin(self):
        return self.__margin

    @property
    def profit(self):
        return self.__profit
    
    @property
    def margin_level(self):
        return self.__margin_level

    @property
    def orders(self):
        return self.__orders

    @property
    def current_candle(self):
        return self.__current_candle

    @current_candle.setter
    def current_candle(self, current_candle: Dict[str, float]):
        for i in  self.__current_candle:
            self.__current_candle[i] = current_candle[i]

    def __init__(self, init_deposit: float = 20000.0):
        self.__open_date = datetime.now(tz=timezone('America/Bogota'))
        self.__id = self.__open_date.strftime('%Y%m%d%H%M%S%f')

        self.__deposits.append(init_deposit)

    def __setattr__(self, name: str, value: Any) -> None:
        super().__setattr__(name, value)

        if name == '_Account__orders':
            self.__update_state()

        if name == 'current_candle':
            self.__update_state()

    def __summarize(self, order: Order, summary: Dict[str, Dict[str, float]]) -> None:
        summary[order.state]['profit'] = summary[order.state]['profit'] + order.profit
        summary[order.state]['margin'] = summary[order.state]['margin'] + order.margin

    def __update_state(self):
        tmp: list = []

        for order_id in self.__orders['pending']:
            self.__orders['pending'][order_id].update_state()
            state: str = self.__orders['pending'][order_id].state
            if state != 'pending':
                self.__orders[state][order_id] = self.__orders['pending'][order_id]
                tmp.append(self.__orders['pending'][order_id].id)

        for order_id in tmp:
            del self.__orders['pending'][order_id]
        
        tmp: list = []

        for order_id in self.__orders['open']:
            self.__orders['open'][order_id].update_state()
            state: str = self.__orders['open'][order_id].state
            if state != 'open':
                self.__orders[state][order_id] = self.__orders['open'][order_id]
                tmp.append(self.__orders['open'][order_id].id)
        
        for order_id in tmp:
            del self.__orders['open'][order_id]

        list(map(lambda order_id:  self.__orders['pending'][order_id].update_state(), self.__orders['pending']))

        summary = {
            'pending': {
                'profit': 0.0,
                'margin': 0.0
            },
            'open': {
                'profit': 0.0,
                'margin': 0.0
            },
            'close': {
                'profit': 0.0,
                'margin': 0.0
            }
        }

        list(map(lambda order_id: self.__summarize(self.__orders['open'][order_id], summary), self.orders['open']))
        list(map(lambda order_id: self.__summarize(self.__orders['close'][order_id], summary), self.orders['close']))

        deposits: float = sum(self.__deposits)

        # Summary
        self.__margin = summary['open']['margin']
        self.__profit = summary['open']['profit']
        self.__balance = deposits + summary['close']['profit']
        self.__heritage = self.__balance + self.__profit
        self.__free_margin = self.__balance - self.__margin
        try:
            self.__margin_level = (self.__heritage / self.__margin) * 100
        except ZeroDivisionError:
            self.__margin_level = 0

    def get_state(self) -> Dict[str, Any]:
        return {
            # Init
            'id': self.__id,
            'deposits': self.__deposits,
            'open_date': self.__open_date,

            # Summary
            'heritage': self.__heritage,
            'balance': self.__balance,
            'free_margin': self.__free_margin,
            'margin': self.__margin,
            'profit': self.__profit,
            'margin_level': self. __margin_level,

            # Current Candles
            'current_candle': self.__current_candle,
        }

    def add_order(self, order: Order) -> None:
        order.current_candle = self.current_candle
        self.__orders['pending'][order.id] = order
        self.__orders = self.__orders

    def __update_order(self, order_id, current_candle):
        self.__orders['pending'][order_id].current_candle = current_candle
        self.__orders = self.__orders

def backtesting(candles: pd.DataFrame, predicts: pd.DataFrame, dir: str, strdatetime: str):
    # Account Create
    account: Account = Account()

    account.current_candle = {
        'timestamp': candles.iloc[0]['timestamp'],
        'open': candles.iloc[0]['open'],
        'high': candles.iloc[0]['high'],
        'low': candles.iloc[0]['low'],
        'close': candles.iloc[0]['close']
    }

    # Orders Create
    for i in range(predicts.shape[0]):
        if predicts.iloc[i]['vf']:
            order: Order = Order(
                predicts.iloc[i]['y_timestamp'],
                'EURUSD',
                0 if predicts.iloc[i]['op'] == 'c' else 1,
                0.01,
                predicts.iloc[i]['stop_loss'],
                predicts.iloc[i]['take_profit']
            )
            account.add_order(order)
    
    states: list = []

    # State Update
    for i in range(candles.shape[0]):
        candle: dict = {
            'timestamp': candles.iloc[i]['timestamp'],
            'open': candles.iloc[i]['open'],
            'high': candles.iloc[i]['high'],
            'low': candles.iloc[i]['low'],
            'close': candles.iloc[i]['close']
        }
        account.current_candle = candle

        state: dict = account.get_state().copy()
        states.append(state)

    # Save
    new_dir = Path(f'{dir}/simulations/{strdatetime}')

    new_dir.mkdir(parents=True, exist_ok=True)

    backtesting: pd.DataFrame = pd.DataFrame()
    for i in account.orders['close']:
        data: dict = account.orders['close'][i].get_state()
        del data['current_candle']
        tmp: pd.DataFrame = pd.DataFrame([data]).set_index('id')
        backtesting = pd.concat([backtesting, tmp])

    states: pd.DataFrame = pd.DataFrame(states)

    backtesting.to_csv(f'{dir}/simulations/{strdatetime}/backtesting.csv')
    states.to_csv(f'{dir}/simulations/{strdatetime}/states.csv')

    return account, backtesting, states
```

### Building

Construcción de red neuronal Perceptrón Multicapa (MLP) con optimización de hiperparámetros.

```python
input_shape: tuple = x_train[0].shape
output_size: tuple = y_train[0].size

def model_builder(hp: kt.HyperParameters) -> kt.HyperModel:
    model = Sequential()

    input = Input(shape=input_shape)
    output = Dense(units=output_size, activation='relu')

    hp_layers_shape = hp.Int('layers_shape', 1, 50, 5)

    model.add(input)

    for i in range(hp_layers_shape):
        model.add(
            Dense(
                units=hp.Int(f'units_{i + 1}', 50, 200, 50),
                activation=hp.Choice(f'activation_type_{i + 1}', ['linear', 'relu'])
            )
        )
            
    model.add(output)
            
    hp_optimizer_type = hp.Choice('optimizer_type', ['SGD', 'RMSprop'])
    hp_learning_rate = hp.Choice('learning_rate', [0.001, 0.005, 0.01, 0.05, 0.1])

    with hp.conditional_scope('optimizer_type', ['SGD']):
        if hp_optimizer_type == 'SGD':
            optimizer = SGD(
                learning_rate=hp_learning_rate
            )
    with hp.conditional_scope('optimizer_type', ['RMSprop']):
        if hp_optimizer_type == 'RMSprop':
            optimizer = RMSprop(
                learning_rate=hp_learning_rate
            )

    model.compile(
        optimizer=optimizer,
        loss='mse',
        metrics=['mse']
    )

    return model
```

# LSTM

Long-Short Term Memory(LSTM) son una extensión de las redes neuronales recurrentes, que básicamente amplían su memoria para aprender de experiencias importantes que han pasado hace mucho tiempo.  Las LSTM permiten a las RNN recordar sus entradas durante un largo período de tiempo. Esto se debe a que LSTM contiene su información en la memoria, que puede considerarse similar a la memoria de un ordenador, en el sentido que una neurona de una LSTM puede leer, escribir y borrar información de su memoria.

### Building

Construcción de red neuronal Long Short Term Memory (LSTM) con optimización de hiperparámetros.

```python
input_shape: tuple = x_train[0].shape
output_size: tuple = y_train[0].size

def model_builder(hp: kt.HyperParameters) -> kt.HyperModel:
    model = Sequential()

    hp_units = hp.Int('units', 50, 200, 50)
    hp_activation = hp.Choice('activation', ['linear', 'relu', 'tanh', 'sigmoid', 'softmax'])
    hp_recurrent_activation = hp.Choice('recurrent_activation', ['linear', 'relu', 'tanh', 'sigmoid', 'softmax'])

    input = Input(shape=input_shape)
    hidden = LSTM(units=hp_units, activation=hp_activation, recurrent_activation=hp_recurrent_activation) 
    output = Dense(units=output_size, activation='relu')

    model.add(input)   
    model.add(hidden)
    model.add(output)
            
    hp_optimizer_type = hp.Choice('optimizer_type', ['SGD', 'RMSprop'])
    hp_learning_rate = hp.Choice('learning_rate', [0.001, 0.005, 0.01, 0.05, 0.1])

    with hp.conditional_scope('optimizer_type', ['SGD']):
        if hp_optimizer_type == 'SGD':
            optimizer = SGD(
                learning_rate=hp_learning_rate
            )
    with hp.conditional_scope('optimizer_type', ['RMSprop']):
        if hp_optimizer_type == 'RMSprop':
            optimizer = RMSprop(
                learning_rate=hp_learning_rate
            )

    model.compile(
        optimizer=optimizer,
        loss='mse',
        metrics=['mse']
    )

    return model
```
